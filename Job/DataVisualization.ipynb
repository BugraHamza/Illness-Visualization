```python
#IMPORT THE LIBRARIES
import csv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#OPEN CSV FILE FUNCTION
def open_csv_file(filename):
    content=[]

    with open(filename, newline='') as csvfile:
        data = csv.reader(csvfile, delimiter=',', quotechar='"')

        for row in data:
            content.append(row)
            
    return content
```

```python
#OPEN covid-like-illness.csv FILE
content1=open_csv_file('covid-like-illness.csv')
```

```python
#INITIALIZE
graphs = [np.array(content1[7+45*i:50+45*i]) for i in range(11)]
names=['National']
names.extend('Region'+str(i) for i in range(1,11))
```

```python
#VISUALIZE AND SAVE THE DATA
i=0
for graph in graphs:
    #GET X AXIS
    weeks1 = graph[2:-1, 0]
    
    #GET Y AXISES
    cli = list(map(float, graph[2:-1, 4]))
    ili = list(map(float, graph[2:-1, 6]))

    #PLOT FIGURE
    plt.figure()
    
    #PLOT CURVES
    plt.plot(weeks1,cli, ili, 'o', linestyle='-')
    
    #NAME THE AXISES
    plt.xlabel('Year/Week')
    plt.ylabel('Total Number of ED Visits')
    plt.xticks(rotation=90)

    #SET LEGEND
    plt.legend(['CLI', 'ILI'])
    
    #SET TITLE
    plt.title(names[i])    
    
    #SAVE THE FIGURE
    plt.savefig(names[i])
    i+=1
```

```python
#OPEN Weekly_Data_Counts.csv FILE
content2=open_csv_file('Weekly_Data_Counts.csv')
```

```python
#INITIALIZE
data=np.array(content2[3:23])
labels=data[:,1]
data=data[:,2:].T
bar_graph=[]
    
#GET X AXIS
weeks2=np.arange(len(labels))

#GET Y AXIS
for i in data:
    bar_graph.append(list(map(float,i)))
```

```python
#PLOT THE BAR GRAPH
plt.figure()
plt.bar(weeks2, bar_graph[0])
plt.xticks(weeks2, labels, rotation=90)

#SET LABELS
plt.xlabel('Year/Week')
plt.ylabel('Case Counts')

for i in range(1, len(bar_graph)):
    plt.bar(weeks2, bar_graph[i], bottom=bar_graph[i-1])

#SET LEGEND
plt.legend(['0-4 yr', '5-17 yr', '18-49 yr', '50-64 yr', '65+ yr'])

#SAVE THE FIGURE
plt.savefig('Week Graph')
```

```python
#OPEN nchs-mortality-report.csv FILE
content3=open_csv_file('nchs-mortality-report.csv')
```

```python
#INITIALIZE
data=np.array(content3[7:])
x=[data[i, 0] + data[i,1] for i in range(len(data))]
graph1=list(map(float, data[:,4]))
graph2=list(map(float, data[:,5]))
graph3=list(map(float, data[:,6]))
```

```python
#PLOT THE CURVES
for g in [graph1, graph2, graph3]:
    plt.plot(x, g)

#SET TITLE
plt.title('Week Graph')

#SET LABELS
plt.xlabel('Year/Week')
plt.ylabel('Pneumonia, Influenza or COVID-19 Deaths')
plt.xticks(rotation=90)

#SET LEGEND
plt.legend(['% of Total', 'Baseline', 'Threshold'])

#SAVE FIGURE
plt.savefig('Week Graph')
```
